# K-Means Clustering

```python
from sklearn.cluster import KMeans
```

Since we know that there are just two categories of colleges here, we set the number of clusters as 2

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(colleges.drop('Private',axis=1))
kmeans.cluster_centers_
```

## Performance Analysis

Here we have the real labels of the data and then we can use them to evaluate the performance of our classifier. First, we  create a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.

```python
def converter(cluster):
    if cluster=='Yes':
        return 1
    else:
        return 0

colleges['Cluster'] = colleges['Private'].apply(converter)
colleges.head()
```

Now, we can use the confustion matrix and classification report to find how the classifier performs

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(colleges['Cluster'],kmeans.labels_))
print(classification_report(colleges['Cluster'],kmeans.labels_))
```
